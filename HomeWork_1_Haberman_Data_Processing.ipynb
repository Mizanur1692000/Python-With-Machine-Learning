# Box-Plot

```python
sns.boxplot(x="status", y="age", data=df, color="green")
plt.show
```

```python
sns.boxplot(x="status",y="nodes", data=df)
plt.show
```

```python
sns.boxplot(x="age", y="nodes", data=df)
plt.show
```

```python
sns.boxplot(x="year", y="status", data=df)
plt.show
```

# Violinplot

```python
sns.violinplot(x="status", y="nodes", data=df)
plt.show
```

```python
sns.violinplot(x="status", y="age", data=df, color="yellow")
plt.show
```

# Scatter-Plot

```python
df.plot(kind="scatter", x="year", y="age")
plt.show
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(df, hue="status",height=7)\
    .map(plt.scatter,"age","year")\
    .add_legend();
plt.show();
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(df, hue="status",height=7)\
    .map(plt.scatter,"nodes","year")\
    .add_legend();
plt.show();
```

```python
sns.set_style("whitegrid");
sns.pairplot(df, hue="status",height=6)

plt.show();
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(df, hue="status",height=7)\
    .map(sns.distplot,"age")\
    .add_legend();
plt.title("Histogram of ages of the patients",fontsize=20)
plt.show();
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(df, hue="status",height=7)\
    .map(sns.distplot,"year")\
    .add_legend();
plt.title("Histogram of operation year of the patients",fontsize=20)
plt.show();
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(df, hue="status",height=7)\
    .map(sns.distplot,"nodes")\
    .add_legend();
plt.title("Histogram of auxilarity nodes detected",fontsize=20)
plt.show();
```

# Pie-Chart

```python
df["status"].value_counts().plot(kind="pie", autopct="%1.2f%%", shadow=True, startangle=45, explode=(0.2,0))
plt.title("Pie Chart")
plt.show()
```
